## Species

```python
df = df.rename(columns={'Species ': 'Species'}) #cambia el nombre de la columna 
df.Species.value_counts()[:-100] #chequea cuántos valores únicos hay
```

```python
df.Species=df.Species.fillna('Unknown')
df.Species=df.Species.str.strip()
df.Species=df.Species.str.replace('"', '')
lista_spec = list(df.Species.unique())
lista_spec.sort()
lista_spec[0:50]
```

```python
def orshark(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bor\b", x)):
        x = "Unknown"
        return x
    else:
        return x
    
df.Species = df.Species.apply(orshark)  
lista_spec = list(df.Species.unique())
lista_spec.sort()
lista_spec[0:50]
```

```python
def whiteshark(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bwhite\b.+\bshark\b", x.lower())):
        x = "White Shark"
        return x
    else:
        return x

def tigershark(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\btiger\b.+\bshark\b", x.lower())):
        x = "Tiger Shark"
        return x
    else:
        return x

def bullshark(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bbull\b.+\bshark\b", x.lower())):
        x = "Bull Shark"
        return x
    else:
        return x
    
def angel(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bangel\b.+\bshark\b", x.lower())):
        x = "Angel Shark"
        return x
    else:
        return x

def basking(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bbasking?\b.+\bshark\b", x.lower())):
        x = "Basking Shark"
        return x
    else:
        return x
    
def blacktip(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bblackt?i?p?.+\bshark", x.lower())):
        x = "Blacktip Shark"
        return x
    else:
        return x
    
def blue(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bblue\b.+\bshark\b", x.lower())):
        x = "Blue Shark"
        return x
    else:
        return x
    
def hammerhead(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bhamm?er\s?head", x.lower())):
        x = "Hammerhead Shark"
        return x
    else:
        return x
    
def mako(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bmako\b.+\bshark\b", x.lower())):
        x = "Mako Shark"
        return x
    else:
        return x
    
def dogfish(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bdog\s?fish", x.lower())):
        x = "Dogfish Shark"
        return x
    else:
        return x

def reef(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\breef\b.+\bshark", x.lower())):
        x = "Reef Shark"
        return x
    else:
        return x
    
def dusky(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bdusky\b.+\bshark", x.lower())):
        x = "Dusky Shark"
        return x
    else:
        return x
                      
df.Species = df.Species.apply(whiteshark)  
df.Species = df.Species.apply(tigershark)
df.Species = df.Species.apply(bullshark)  
df.Species = df.Species.apply(angel) 
df.Species = df.Species.apply(basking)  
df.Species = df.Species.apply(blacktip)
df.Species = df.Species.apply(blue)  
df.Species = df.Species.apply(hammerhead) 
df.Species = df.Species.apply(mako)  
df.Species = df.Species.apply(dogfish) 
df.Species = df.Species.apply(reef) 
df.Species = df.Species.apply(dusky)  


lista_spec = list(df.Species.unique())
lista_spec.sort()

lista_spec[0:50]
```

```python
df.Year = df.Year.astype(str) 
df.Year=df.Year.str.strip()
df.Year=df.Year.str.replace('Reported','')
df.Year=df.Year.str.replace('January','Jan')
df.Year=df.Year.str.replace('February','Feb')
df.Year=df.Year.str.replace('March','Mar')
df.Year=df.Year.str.replace('April','Apr')
df.Year=df.Year.str.replace('May','May')
df.Year=df.Year.str.replace('June','Jun')
df.Year=df.Year.str.replace('July','Jul')
df.Year=df.Year.str.replace('August','Aug')
df.Year=df.Year.str.replace('September','Sep')
df.Year=df.Year.str.replace('October','Oct')
df.Year=df.Year.str.replace('November','Nov')
df.Year=df.Year.str.replace('December','Dec')
df.Year=df.Year.str.strip()

def year(x): #si la fecha tiene formaro de dd-mmm-aa devuelve únicamente el año
    if re.search(r"^\d?\d-\w\w\w-(\d\d\d?\d?)$", x.lower()):
        ye = re.search(r"^\d?\d-\w\w\w-(\d\d\d?\d?)$", x.lower())
        x = ye.group(1)
        return x
    else:
        return x
                      
df.Year = df.Year.apply(year) 
```

```python
df.Activity = df.Activity.fillna("Unknown")
df.Activity=df.Activity.str.strip()
df.Activity = df.Activity.str.replace('"','')
```

```python
def surfing(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bsurf", x.lower())):
        x = "Surfing"
        return x
    else:
        return x

def swimming(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bswim", x.lower())):
        x = "Swimming"
        return x
    else:
        return x

def fishing(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bfish", x.lower())):
        x = "Fishing"
        return x
    else:
        return x
    
def wading(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bwad", x.lower())):
        x = "Wading"
        return x
    else:
        return x

def bathing(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bbath", x.lower())):
        
        x = "Bathing"
        return x
    else:
        return x
    
def standing(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bstand", x.lower())):
        x = "Standing"
        return x
    else:
        return x
    
def scuba(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bs[cko]uba", x.lower())):
        x = "Scuba Diving"
        return x
    else:
        return x
    
def diving(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bdeep\sdiving", x.lower())):
        x = "Deep Diving"
        return x
    else:
        return x

def diving2(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"^div[ie]", x.lower())):
        x = "Diving"
        return x
    else:
        return x
    
def snorkeling(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bsnorkel", x.lower())):
        x = "Snorkeling"
        return x
    else:
        return x
    
def bodyboarding(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bbody[\s-]?boarding", x.lower())):
        x = "Body Boarding"
        return x
    else:
        return x

def accident1(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\baccident", x.lower())):
        x = "Accident"
        return x
    else:
        return x

def accident2(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bf[ea]ll", x.lower())):
        x = "Accident"
        return x
    else:
        return x
    
    
def accident3(x): #si la string tiene or devuelve unknown
    if bool(re.search(r"\bdisaster", x.lower())):
        x = "Accident"
        return x
    else:
        return x

def otheractivity(x): #si la string tiene or devuelve unknown
    if x != "Surfing" and x != "Swimming" and x != "Fishing" and x != "Wading" and x != "Bathing" and x != "Standing" and x != "Scuba Diving" and x != "Deep Diving" and x != "Diving" and x != "Snorkeling" and x != "Body Boarding" and x != "Accident":
        x = "Other Activity"
        return x
    else:
        return x

                      
df.Activity = df.Activity.apply(surfing)  
df.Activity = df.Activity.apply(swimming)
df.Activity = df.Activity.apply(fishing)  
df.Activity = df.Activity.apply(wading) 
df.Activity = df.Activity.apply(bathing)  
df.Activity = df.Activity.apply(standing)
df.Activity = df.Activity.apply(diving) 
df.Activity = df.Activity.apply(diving2)  
df.Activity = df.Activity.apply(snorkeling) 
df.Activity = df.Activity.apply(scuba)  
df.Activity = df.Activity.apply(bodyboarding) 
df.Activity = df.Activity.apply(accident1) 
df.Activity = df.Activity.apply(accident2)  
df.Activity = df.Activity.apply(accident3) 
df.Activity = df.Activity.apply(otheractivity)  


lista_act = list(df.Activity.unique())
lista_act.sort(reverse=True)
lista_act
```
